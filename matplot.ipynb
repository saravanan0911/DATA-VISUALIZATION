```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
plt.plot(a,b,'y+--', linewidth=1 , markersize=10)
plt.title('line plot', size='25')
plt.xlabel('x-axis', size='10')
plt.ylabel('y-axis', size='10')
```

 #### QUESTION 2

```python
d = [1,2,3,4,5,6,7] 
s1 = [160,150,140,145,175,165,180] 
s2 = [70,90,160,150,140,145,175] 
plt.title('sales vs days', size='25')
plt.xlabel('days', size='15')
plt.ylabel('sales', size='15')
plt.plot(d,s1,'y+--',linewidth=2,markersize=10,label='sales of company 1')
plt.plot(d,s2,'b*--',linewidth=2,markersize=10,label='sales of company 2')
plt.legend(loc='upper left')
```
